```python
import pandas as pd
import re
import string
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
df = pd.read_excel("reviews_full (1).xlsx")
print("Jumlah total ulasan:", len(df))
print("\nContoh 5 data awal:")
display(df.head())
```

Case Folding

```python
df['casefolded'] = df['review'].astype(str).str.lower()
print("\n✅ HASIL CASE FOLDING (huruf kecil):")
display(df[['review', 'casefolded']].head())
```

CLEANING

```python
def clean_text(text):
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'@\w+|#\w+', '', text)
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'\s+', ' ', text)
    text = text.strip()
    return text

df['cleaned'] = df['casefolded'].apply(clean_text)
print("\n✅ HASIL CLEANING (hapus tanda baca, angka, simbol):")
display(df[['casefolded', 'cleaned']].head())
```

Tokenizing

```python
nltk.download('punkt_tab')
df['tokenized'] = df['cleaned'].apply(word_tokenize)
print("\n✅ HASIL TOKENIZING (kalimat dipisah jadi kata):")
display(df[['cleaned', 'tokenized']].head())
```

FILTERING

```python
stop_words = set(stopwords.words('indonesian'))
df['filtered'] = df['tokenized'].apply(lambda tokens: [w for w in tokens if w not in stop_words])
print("\n✅ HASIL FILTERING (hapus kata umum):")
display(df[['tokenized', 'filtered']].head())
```

```python
STEMMING
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()
df['stemmed'] = df['filtered'].apply(lambda tokens: [stemmer.stem(w) for w in tokens])
print("\n✅ HASIL STEMMING (ubah kata ke bentuk dasar):")
display(df[['filtered', 'stemmed']].head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

df['stemmed_text'] = df['stemmed'].apply(lambda tokens: ' '.join(tokens))

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['stemmed_text'])

tf = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
print("Matriks Term Frequency:")
tf.head()
```

```python
df['preprocessed'] = df['stemmed'].apply(lambda tokens: ' '.join(tokens))
print("\n✅ HASIL AKHIR (teks siap analisis):")
display(df[['review', 'preprocessed']].head())
```

```python
df[['review', 'preprocessed']].to_csv("bhumi_merapi_preprocessed.csv", index=False, encoding="utf-8-sig")
print("\n💾 File hasil preprocessing disimpan sebagai: 'bhumi_merapi_preprocessed.csv'")
```
